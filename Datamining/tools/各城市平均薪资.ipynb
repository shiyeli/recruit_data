```python
from tools import tool
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#获取城市，薪资
sql="""
	SELECT city,salary_high,salary_low FROM zhilian_job;
"""
city_salary_data=pd.read_sql(sql,tool.get_connnect())

#数据预处理
city_salary_data['salary_average']=city_salary_data['salary_high']+city_salary_data['salary_low']
city_salary_data.drop(['salary_high','salary_low'],axis=1,inplace=True)
city_salary_data.tail()
```

```python
grouped_des=city_salary_data.groupby('city').describe()
res_filter=grouped_des[grouped_des['salary_average']['count']>200]
res_sorted=res_filter.sort_values(('salary_average','50%'),ascending=False)
res_handled=res_sorted.reset_index()
res_handled.head()
```

```python
res_median=pd.concat([res_handled['city'],res_handled[('salary_average','50%')]],axis=1).rename(columns={('salary_average','50%'):'salary_median'})
res_median.head()
```

```python
res_std=pd.concat([res_handled['city'],res_handled[('salary_average','std')]],axis=1).rename(columns={('salary_average','std'):'std'})
res_std.head()
```

```python
#使用echars展示，此处只做数据导出，格式为{name:'成都',value:35}
#http://echarts.baidu.com/examples/editor.html?c=map-polygon
#http://gallery.echartsjs.com/editor.html?c=heatmap-map

res_median['salary_median']=res_median['salary_median'].apply(lambda x:int(x))
median_json=res_median.rename(columns={'city':'name','salary_median':'values'}).to_json(orient='records',force_ascii=False)
median_json
```

```python
res_std['std']=res_std['std'].apply(lambda x:int(x))
std_json=res_std.rename(columns={'city':'name','std':'values'}).to_json(orient='records',force_ascii=False)
std_json
```
